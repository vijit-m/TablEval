```
import pandas as pd 
with open("/content/test.txt", 'r') as f:
  k=f.read()
  k1=k.split("id: ")
  k2=k.split("original_text: ")
# print(len(k1))
l1=[]
l2=[]
for i in range(1,len(k1)):
  ko=k1[i].split("\n")
  ko=ko[0]
  ko=ko[1:-1]
  l1.append(ko)
print(len(k2))
for j in range(1,len(k2)):
  kk=k2[j].split("\n")
  kk=kk[0]
  kk=kk[1:-1]
  l2.append(kk)
# print(l2)
```

```
df = pd.DataFrame(list(zip(l1, l2)), 
               columns =['Statement_id', 'Statement']) 
df.to_csv("/content/test.csv",mode = 'w', index=False, index_label=False)
```
